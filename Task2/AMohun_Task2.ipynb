```python
from pyspark import SparkContext
sc = SparkContext(appName="StormApp").getOrCreate()
```

```python
rdd = sc.textFile("/Users/courtneymohun/Downloads/flood_data_small.csv", minPartitions=None)
```

```python
rdd.collect()
```

```python
header = rdd.first()
rdd = rdd.filter(lambda x: x != header)
rdd = rdd.map(lambda x: (x.split(",")))
```

```python
header=header.split(",")
```

```python
directinjuries=header.index("INJURIES_DIRECT")
indirectinjuries=header.index("INJURIES_INDIRECT")
directdeaths=header.index("DEATHS_DIRECT")
indirectdeaths=header.index("DEATHS_INDIRECT")
propertydamage=header.index("DAMAGE_PROPERTY")
cropdamage=header.index("DAMAGE_CROPS")
mycolumns = [directinjuries, indirectinjuries, directdeaths, indirectdeaths, propertydamage, cropdamage]

for column in mycolumns:
    print(f"x[{column}] != '' AND")
```

```python
rdd_filter = rdd.filter(lambda x: x[26] != '' and x[22] != '' and x[23] != '' and x[24] != '' and x[25] != '' and x[27] != '')
rdd_filter = rdd_filter.map(lambda x: (x[14],  x[26][:-1]))

rdd_filter.collect()

```

```python
header = rdd_filter.first()
rdd_filter3 = rdd_filter.filter(lambda x: x != header)
rdd_filter3 = rdd_filter.filter(lambda x: x != header)

rdd_filter3.collect()
```

```python
from statistics import mean
rdd_organized = rdd_filter3.reduceByKey(lambda a, b: mean([float(a), float(b)]))
rdd_organized.collect()

```
